```python
from matplotlib import pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
tr_data = pd.read_csv('train_logs.csv')
ts_data = pd.read_csv('test_logs.csv')
```

```python
plt.plot(tr_data.train_loss)
plt.title('Train Loss')
```

```python
plt.plot(ts_data.test_accuracy)
plt.title('Test Accuracy')
```

```python
plt.plot(tr_data.grad_norm)
plt.title('Gradient Magnitude ($\ell_2$)')
```

```python
tr_log_tpl = 'lr0.001_bs256_wd{}_connx0_train_logs.csv'

plt.figure()

for wd in ['0', '1e-05', '1e-06', '1e-07']:
    if wd == '0':
        data = pd.read_csv('train_logs.csv')
    else:
        data = pd.read_csv(tr_log_tpl.format(wd))
    plt.plot(data.train_loss, label=f'weight_decay={wd}')

    
plt.xlabel('Epoch')
plt.ylabel('Train loss')
plt.legend()
plt.title('Impact of weight decay on training loss')
```

```python
tr_log_tpl = 'lr0.001_bs256_wd{}_connx0_test_logs.csv'

plt.figure()

for wd in ['0', '1e-05', '1e-06', '1e-07']:
    if wd == '0':
        data = pd.read_csv('test_logs.csv')
    else:
        data = pd.read_csv(tr_log_tpl.format(wd))
    plt.plot(data.test_accuracy, label=f'weight_decay={wd}')

    
plt.xlabel('Epoch')
plt.ylabel('Test accuracy')
plt.legend()
plt.title('Impact of weight decay on test set accuracy')
```

```python
tr_log_tpl = 'lr0.001_bs256_wd0_connx0_t{}_train_logs.csv'

plt.figure()

for bw in ['0.05', '0.01', '0.025', '0.005']:
    if bw == '0.05':
        data = pd.read_csv('train_logs.csv')
    else:
        data = pd.read_csv(tr_log_tpl.format(bw))
    plt.plot(data.train_loss, label=f't_bin_width={bw} s')

    
plt.xlabel('Epoch')
plt.ylabel('Train loss')
plt.legend()
plt.title('Impact of time bin width on training loss')

ts_log_tpl = 'lr0.001_bs256_wd0_connx0_t{}_test_logs.csv'

plt.figure()

for bw in ['0.05', '0.01', '0.025', '0.005']:
    if bw == '0.05':
        data = pd.read_csv('test_logs.csv')
    else:
        data = pd.read_csv(ts_log_tpl.format(bw))
    plt.plot(data.test_accuracy, label=f't_bin_width={bw} s')

    
plt.xlabel('Epoch')
plt.ylabel('Test accuracy')
plt.legend()
plt.title('Impact of time bin width on test set accuracy')
```
